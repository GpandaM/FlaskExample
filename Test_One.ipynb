```python
fig, ax = plt.subplots(1,2)
sns.regplot('RM', 'MEDV', df, ax = ax[0], scatter_kws={'alpha':0.3})
sns.regplot('LSTAT', 'MEDV', df, ax = ax[1], scatter_kws={'alpha':0.4})
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,8))
fig
ax[0] = sns.residplot('RM', 'MEDV', df, ax = ax[0], scatter_kws = {'alpha':0.3})
ax[0].set_ylabel('MEDV residulas $(y-\hat{y})$')
ax[1] = sns.residplot('LSTAT', 'MEDV', df, ax = ax[1], scatter_kws = {'alpha':0.3})
ax[1].set_ylabel('')
```
